# Lowercase, remove new line symbols

```python
df['message'][100]
```

```python
df['message_lowercase'] = df['message'].str.lower()
newLine ="\\n|\\r"
df['message_lowercase'] = df['message_lowercase'].str.replace(newLine,' ',regex=True).str.strip()
df.head()
```

```python
df.to_csv('./../labeledTweets/allLabeledTweets_clean.csv', index=False)
```
